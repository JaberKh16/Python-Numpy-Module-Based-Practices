# Array Creation using __new__ magic method


### Here,2 modes is there to create an array
   
   '''<br>
       1. When no buffere is there, shape , dtype, order is used. <br>
       2. When buffer is there,means buffer is exposed using buffer interface ,
          then all keywords are      interpreted  
   '''

```python
### when buffer = None
try :
    x = np.array(shape = (2, 2), dtype = float, order = 'F')
    x
except Exception as e :
    print(e)
```

```python
### when buffer is active
try :
    x = np.array(shape = (2,), buffer = np.array([1, 2, 3]), offset = 1 , dtype = int)
    x
    
except Exception as e :
    print(e)
```

### Repetance over the array using repeat(repeat[,axis])

```python
x = np.array([[[1, 59, 2, 12], [23,4, 2, 20], [23, 4, 6, 86]]])
print(x)
print("-------------------------------------------------")
print(x.repeat(1)) # this will repeat the array every elements in array as many time as we wanted, here we 1 time
print("--------------------------------------------------------------")
print(x.repeat(3)) # here we want the element to repeat 3 times
```

```python
x = np.array([[[1, 59, 2, 12], [23,4, 2, 20], [23, 4, 6, 86]]])
print(x)
print("--------------------------------")

print(x.repeat(1, axis = 1) ) # this will repeat the array column elements in array as many time as we wanted, here,1 time
print("--------------------------------------------------------------")
print(x.repeat(1, axis = 0))  # this will repeat the array row elements in array as many time as we wanted, here,1 time
print("--------------------------------------------------------------")


```

```python
x = np.array([[[1, 59, 2, 12], [23,4, 2, 20], [23, 4, 6, 86]]])
print(x)
print("--------------------------------")

# repeating 3 times over axis = 0 and axis = 1
print(x.repeat(3, axis = 0)) # reapting 3 times over the column(horizontal fashion)
print("--------------------------------------------------------------")

```

```python
x = np.array([[[1, 59, 2, 12], [23,4, 2, 20], [23, 4, 6, 86]]])
print(x)
print("--------------------------------")

# repeating 3 times over axis = 0 and axis = 1
print(x.repeat(3, axis = 1)) # reapting 3 times over the rows (vertical fashion)
print("--------------------------------------------------------------")
```

# General Broadcasting Rules

'''
    
    When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions,
    and works its way forward. Two dimensions are compatible when --->
                                 
                                  1. they are equal, or
                                  2. one of them is 1
                                  
   
    If these conditions are not met, a ValueError: operands could not be broadcast together exception is thrown, 
    indicating that the arrays have incompatible shapes. The size of the resulting array is the maximum size along 
    each dimension of the input arrays.

'''

### #3 Array Newaxis Workism For Broadcasting

```python
A = np.array([3, 11, 4, 5])
B = np.array([5, 0, 3])

# broadcasting using newaxis here if row = np.newaxis then rows append under [[rows]] or if column = np.newaxis then column append under [[column]]
A = A[np.newaxis, :]
B = B[:, np.newaxis]
C = A- B
C
    
```
